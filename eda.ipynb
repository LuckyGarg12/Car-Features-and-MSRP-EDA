```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/data.csv')
df
```

```python
df.info()
```

#### Summary Statistics 
* Generates summary statistics for all columns in the DataFrame, including count, mean, standard deviation, min, max, and quartiles.
* Prints the summary statistics.

#### Additional Statistical Calculations
* Identifies numerical columns in the DataFrame.
* For each numerical column, calculates and prints the mean, median, mode, variance, and standard deviation.


##### Inference:
* Graph 1 -> The count of new cars has increased greatly in the recent years.
* Graph 2 -> Majority of the cars have horsepower in the range of 100-400.
* Graph 3 -> Most cars have a 4,6 or 8 cylinders. 5, 10, 12 cylinder's are extremely rare to find.
* Graph 4 -> Most cars have 4 doors followed by 2 doors, few cars also have 3 doors.
* Graph 5 -> The Highway MPG of most cars is between 25 to 45
* Graph 6 -> The City MPG of of most cars is between 15 to 25
* Graph 7 -> Most cars have popularity between 100 to 2000, some also have popularity of 5000
* Graph 8 -> Most cars have price between 10000 and 30000

```python
summary_stats = df.describe(include='all')
print(summary_stats)


numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    print(f"{col} - Mean: {df[col].mean()}, Median: {df[col].median()}, Mode: {df[col].mode()[0]}, Variance: {df[col].var()}, Std Dev: {df[col].std()}")


categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"{col} unique values: {df[col].nunique()}")


for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### Boxplots
* Show the distribution of data.
* Highlight the median, quartiles, and potential outliers.
* Provide a visual summary of the data's central tendency and variability.

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

#### Correlation Matrix
* Shows the strength and direction of relationships between numerical columns.
* Helps identify highly correlated features which may impact model performance.

#### Pairplot:
* Visualizes pairwise relationships and distributions of selected numerical columns.
* Helps identify patterns, trends, and potential outliers in the data.
* Creates a pairplot for selected numerical columns: 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', and 'MSRP'.

#### Inference
* Engine HP has very high correlation with Engine cylinders
* Engine HP and MSRP also have very high corelation

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

sns.pairplot(numeric_df[['Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'MSRP']])
plt.show()
```

#### Trend Analysis
* Visualizes the trend of average car prices over the years.
* Helps identify patterns, such as increases or decreases in car prices over time.

#### Inference
* There has been considerable spike in prices in 2000 and 2008.

```python
# Trend Analysis
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Year'], y=df['MSRP'])
plt.title('Average Car Price Over Years')
plt.show()
```

#### Fuel Efficiency Analysis
* Visualizes the distribution of highway MPG for different engine fuel types.
* Helps identify which fuel types are associated with higher or lower fuel efficiency on highways.
* Highlights the variability and potential outliers in highway MPG for each fuel type.

#### Inference
* Electric Fuel gives the highest Highway MPG

```python
# 5. Fuel Efficiency Analysis
plt.figure(figsize=(8, 6))
sns.boxplot(x='Engine Fuel Type', y='highway MPG', data=df)
plt.xticks(rotation=45)
plt.title('Fuel Type vs Highway MPG')
plt.show()
```

#### Price Analysis
* Identifies the car makes with the highest average prices.
* Provides insights into which car brands are positioned as premium or luxury based on their average MSRP.
* Helps in understanding market positioning and pricing strategies of different car manufacturers.
#### Inference
* Bugatti cars are the most costliest cars costing more than done double of the other brands

```python
# 6. Price Analysis
plt.figure(figsize=(8, 6))
top_makes = df.groupby('Make')['MSRP'].mean().nlargest(10)
sns.barplot(x=top_makes.index, y=top_makes.values)
plt.xticks(rotation=45)
plt.title('Top 10 Car Makes by Average Price')
plt.show()
```
